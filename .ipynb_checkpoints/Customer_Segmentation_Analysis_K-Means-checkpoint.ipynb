### Impotando Librerías

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 

sns.set() 
```

### Exploración de los datos

```python
raw_data = pd.read_csv('Mall_Customers.csv')
raw_data.head()
```

```python
raw_data.describe()
```

```python
raw_data.isnull().sum()
```

```python
raw_data.dtypes
```

```python
# Trabajamos con un respaldo
data = raw_data.copy()

# Convertir la variable categórica "Gender" a valores numéricos para facilitar el análisis estadístico.
data['Gender'] = data['Gender'].map({'Male': 0, 'Female': 1})
# Dropear la columna CustomerID ya que no aporta información relevante
data = data.drop(columns=['CustomerID'])

data.head()
```

```python
# Histograma para ver la distribución de cada variable numérica
data.hist(figsize=(10, 6), bins=20)
plt.show()
```

```python
plt.figure(1 , figsize = (15 , 6))
n = 0 
for x in ['Age' , 'Annual Income (k$)' , 'Spending Score (1-100)']:
    n += 1
    plt.subplot(1 , 3 , n)
    plt.subplots_adjust(hspace =0.5 , wspace = 0.5)
    sns.distplot(data[x], bins=20, kde=True)
    plt.title('Distplot of {}'.format(x))
plt.show()
```
